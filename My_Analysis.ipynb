# Binarized data

```python
column_names = [
        'top-left-square', 'top-middle-square', 'top-right-square',
        'middle-left-square', 'middle-middle-square', 'middle-right-square',
        'bottom-left-square', 'bottom-middle-square', 'bottom-right-square',
        'Class'
    ]
df = pd.read_csv('/content/FCALC/data_sets/tic-tac-toe.data', names = column_names)
df['Class'] = [x == 'positive' for x in df['Class']]
df.head()
```

```python
X = pd.get_dummies(df[column_names[:-1]], prefix=column_names[:-1]).astype(bool)
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_test
```

```python

bin_cls = fcalc.classifier.BinarizedBinaryClassifier(X_train.values, y_train.to_numpy(), method="standard-support")
```

```python
bin_cls.predict(X_test.values)
```

```python
X_test
```

```python
from sklearn.metrics import accuracy_score, f1_score

print(accuracy_score(y_test, bin_cls.predictions))
print(f1_score(y_test, bin_cls.predictions))
```

$\textbf{Donwloading datasets}$

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
heart_disease = fetch_ucirepo(id=45)
cars = fetch_ucirepo(id=19)
bank_subscription = fetch_ucirepo(id = 222)


X_heart_disease = heart_disease.data.features
y_heart_disease = heart_disease.data.targets
y_heart_disease = pd.DataFrame(y_heart_disease)
X_heart_disease = pd.DataFrame(X_heart_disease)
X_bank_subscription = bank_subscription.data.features
y_bank_subscription = bank_subscription.data.targets

df = pd.concat([X_heart_disease, y_heart_disease], axis = 1)
df=df.dropna(axis=0)


y_heart_disease = df.iloc[:,-1:]
X_heart_disease = df.iloc[:,:-1]
y_heart_disease=y_heart_disease.apply(lambda x: (x>0)*1)
X_cars = cars.data.features
y_cars = cars.data.targets


y_cars.value_counts()
```

#$\textbf{RandomForest classification for DataSets}$

$\textbf{RandomForest classification for Heart Diseases}$

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 100)
X_train_heart_disease, X_test_heart_disease, y_train_heart_disease, y_test_heart_disease = train_test_split(X_heart_disease, y_heart_disease, test_size=0.3, random_state=42)
clf.fit(X_train_heart_disease, y_train_heart_disease)



y_pred_heart_disease = clf.predict(X_test_heart_disease)
```

```python
fpr, tpr, thresholds = roc_curve(y_test_heart_disease, y_pred_heart_disease.reshape(90,1))
roc_auc = auc(fpr, tpr)

print("recall: ", recall_score(y_test_heart_disease, y_pred_heart_disease.reshape(90,1)))
print("precision: ",precision_score(y_test_heart_disease, y_pred_heart_disease.reshape(90,1)) )
print("accuracy: ",accuracy_score(y_test_heart_disease, y_pred_heart_disease.reshape(90,1)))
print("f1 score: ",f1_score(y_test_heart_disease, y_pred_heart_disease.reshape(90,1)))
print("ROC AUC:", roc_auc)
data = [[recall_score(y_test_heart_disease, y_pred_heart_disease.reshape(90,1)),precision_score(y_test_heart_disease, y_pred_heart_disease.reshape(90,1)),
         accuracy_score(y_test_heart_disease, y_pred_heart_disease.reshape(90,1)),f1_score(y_test_heart_disease, y_pred_heart_disease.reshape(90,1)),roc_auc]]
df_stat = pd.DataFrame(data, columns=['recall', 'precision','accuracy','f1 score','ROC AUC'])
```
